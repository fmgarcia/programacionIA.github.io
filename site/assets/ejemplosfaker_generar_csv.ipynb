```python
from faker import Faker
import csv

faker = Faker("es_ES")

with open('ficheros/users_fake.csv', 'w', newline='') as f:

    fieldnames = ['id', 'nombre', 'apellidos', 'trabajo']
    writer = csv.DictWriter(f, fieldnames=fieldnames, delimiter=',', quotechar='"')
    writer.writeheader()

    for i in range(1, 1001, 1):
        _id = i
        fname = faker.first_name()
        lname = faker.last_name()
        occupation = faker.job()

        writer.writerow({'id': _id, 'nombre': fname, 
            'apellidos': lname, 'trabajo': occupation})
```
