```python
import json

# Leyendo el archivo profesor.json que contiene un json en formato string y lo paso a un string en Python
with open('ficheros/profesor.json') as f:
    json_content = f.read()
    
# a partir de un Json string lo convertimos a un diccionario de Python
profesor = json.loads(json_content)
print(type(profesor))

print(profesor['nombre']) # Fran
print(profesor['edad']) # 25
print(profesor['domicilio']) # {'direccion': 'Lillo Juan, 128', 'ciudad': 'San Vicente del Raspeig', 'comunidad': 'Valenciana', 'codigoPostal': '03690'}
print(profesor['domicilio']['ciudad']) # San Vicente del Raspeig
print(profesor['numerosTelefonos']) # [{'tipo': 'casa', 'numero': '666 666 666'}, {'tipo': 'movil', 'numero': '777 777 777'}]
print(profesor['numerosTelefonos'][1]['numero']) # 777 777 777

nombre = profesor['nombre']
print(nombre) # Fran
```

```python
# Leer Json desde Internet
import requests # hacer peticiones a internet (pip install requests)
import json

url = "https://jsonplaceholder.typicode.com/todos"
response = requests.get(url)
response_json = response.json() # convierte el json en una lista de Python
print(response_json)
print(type(response_json)) # <class 'list'>
lista_pendientes = [tarea for tarea in response_json if not tarea['completed']]
lista_completadas = [tarea for tarea in response_json if tarea['completed']]
print(len(lista_pendientes)) # 110
print(len(lista_completadas)) # 90
```

```python
# Dada la clase Personaje y su Decoder, leer el json desde internet y cargarlo en un objeto de tipo Personaje
from entidades.personaje import PersonajeDecoder
import requests
import json

def imprimir_personaje(personaje):
    print(f"Nombre: {personaje.nombre}")
    print(f"Altura: {personaje.altura}")
    for pelicula in personaje.peliculas:
        respuesta_pelicula = requests.get(pelicula)
        json_pelicula = json.loads(respuesta_pelicula.text)
        print(f"Pelcula: {json_pelicula['title']} - Director: {json_pelicula['director']}")
    for especie in personaje.especies:
        respuesta_especie = requests.get(especie)
        json_especie = json.loads(respuesta_especie.text)
        print(f"Especie: {json_especie['name']}")
    

codigo = input("Introducte el cdigo del personaje: ")

url= f"https://swapi.dev/api/people/{codigo}/?format=json"
response=requests.get(url)
personaje=json.loads(response.text,cls=PersonajeDecoder)
imprimir_personaje(personaje)
```
