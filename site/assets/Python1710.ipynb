```python
import requests
import json
from entidades.videojuego import Videojuego
from datetime import datetime

key = "08c18ae30e654ff499c546879ec660f2"
videojuegos = []
for id_pagina in range(1, 2):  # En este for obtenemos los datos
    url= f"https://api.rawg.io/api/games?key={key}&page={id_pagina}"
    cabeceras = {}
    cabeceras["User-Agent"] = "PostmanRuntime/7.26.8"
    response=requests.get(url,cabeceras)
    json_data = json.loads(response.text)
    resultados = json_data["results"]  # lista de diccionarios, donde cada diccionario es un juego    
    for juego in resultados: # En este for creamos los objetos
        #print(f"{juego['name']} - Fecha lanzamiento: {juego['released']} - Imagen: {juego['background_image']} - Rating: {juego['rating']}")
        videojuego = Videojuego(juego['name'], datetime.strptime(juego['released'], "%Y-%m-%d").year, juego['background_image'], juego['rating'])
        videojuegos.append(videojuego)
        
# Operaciones con listas (aquí podríamos hacer cualquier operación que se nos ocurra)
datos_csv = videojuegos  # Copia la lista de videojuegos
# estas líneas de código son solo un ejemplo de lo que se puede hacer con las listas
datos_csv = [videojuego for videojuego in videojuegos if videojuego.valoracion > 4]  # Filtra los videojuegos con rating mayor a 4.5
datos_csv.sort(key=lambda x: x.valoracion, reverse=True)  # Ordena la lista de videojuegos por rating de mayor a menor        
           
# Escritura de los datos en un fichero csv. Creo el dataset con los datos de los videojuegos
with open('ficheros/videojuegos.csv', 'w') as f:
    f.write("Nombre, Fecha de lanzamiento, Imagen, Rating\n")  # Imprime cabecera del csv
    for videojuego in datos_csv:  # Imprime los datos de los videojuegos
        f.write(str(videojuego) + "\n")
```
