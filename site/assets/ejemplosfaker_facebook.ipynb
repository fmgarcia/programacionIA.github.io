```python
from faker import Faker
import csv

faker = Faker(['es_ES','fr_FR','en_US','ja_JP'])
cantidad_usuarios = 1000
ruta_fichero = 'ficheros/users_facebook.csv'
campos=["id","nombre","apellidos","correo","password","dia","mes","anyo","genero"]

with open(ruta_fichero, 'w', newline='') as csvfile:
    csvwriter = csv.DictWriter(csvfile, fieldnames=campos , delimiter=',', quotechar='"')
    csvwriter.writeheader()
    for i in range(1, cantidad_usuarios+1, 1):
        perfil = faker.simple_profile()
        csvwriter.writerow(
            {
                "id": i,
                "nombre": perfil["name"].split()[0],               
                "apellidos": perfil["name"].split()[0],
                "correo": perfil["mail"],
                "password": faker.password(length=10,special_chars=True,digits=True),
                "dia": str(perfil["birthdate"].day),
                "mes": perfil["birthdate"].month,
                "anyo": perfil["birthdate"].year,
                "genero": perfil["sex"],
            }
        )
print("Proceso finalizado correctamente...")
```
